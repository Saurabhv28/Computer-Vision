
1. HOG focuses on the structure of the object. It extracts the information of the edges magnitude as well as the orientation of the edges.

2. It uses a detection window of 64x128 pixels, so the image is first converted into (64, 128) shape.

3. The image is then further divided into small parts, and then the gradient and orientation of each part is calculated. It is divided into 8x16 cells into blocks with 50% overlap, so there are going to be 7x15 = 105 blocks in total, and each block consists of 2x2 cells with 8x8 pixels.

4. We take the 64 gradient vectors of each block (8x8 pixel cell) and put them into a 9-bin histogram.

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from  skimage import exposure
import matplotlib.pyplot as plt
```

```python
 img = imread('/content/drive/MyDrive/102_5495.JPG')
#  plt.axis('off')
 plt.imshow(img)
print(img.shape)
```

```python
import cv2
img2 = img[0:2448, 500:2700]
plt.imshow(img2)
print(img2.shape)
```

```python
resized_img = resize(img2, (128, 64))
plt.axis('off')
plt.imshow(resized_img)
print(resized_img.shape)
```

```python
#creating hog features
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                	cells_per_block=(2, 2), visualize=True, multichannel=True)
plt.axis("off")
plt.imshow(hog_image, cmap="gray")  
```
